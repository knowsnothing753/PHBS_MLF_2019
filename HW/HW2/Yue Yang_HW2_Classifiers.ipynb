# Homework 2: classification
Data source: http://archive.ics.uci.edu/ml/datasets/Polish+companies+bankruptcy+data

```python
from scipy.io import arff
import pandas as pd
import numpy as np
```

```python
data = arff.loadarff('4year.arff')
df = pd.DataFrame(data[0])
```

```python
df['bankruptcy'] = (df['class']==b'1')
del df['class']
df.columns = ['X{0:02d}'.format(k) for k in range(1,65)] + ['bankruptcy']
```

```python
df.describe()
```

```python
sum(df.bankruptcy == True)
```

```python
from sklearn.impute import SimpleImputer
imp_mean = SimpleImputer(missing_values=np.nan, strategy='mean')
X_imp = imp_mean.fit_transform(df.values)
```

```python
from sklearn.model_selection import train_test_split

X, y = X_imp[:, :-1], X_imp[:, -1]

X_train, X_test, y_train, y_test =\
    train_test_split(X, y, 
                     test_size=0.3, 
                     random_state=0, 
                     stratify=y)
```

```python
from sklearn.preprocessing import StandardScaler

stdsc = StandardScaler()
X_train_std = stdsc.fit_transform(X_train)
X_test_std = stdsc.transform(X_test)
```

## Find the 2 most important features
using Logistic Regression with L1 penalty

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
# and put more...
```

```python
lr1 = LogisticRegression(penalty='l1',solver='liblinear',C=0.01) # complete
lr1.fit(X_train_std,y_train)
```

```python
lr1.coef_[lr1.coef_!=0].shape # check the number of the features with non-zero weights
#X1 and X38 are the selected features
```

```python
# redefine X_train_std and X_test_std
X_train_std = X_train_std[:, lr1.coef_[0]!=0]
X_test_std = X_test_std[:, lr1.coef_[0]!=0]
print(X_train_std.shape)
print(X_test_std.shape)
```

## Apply LR / SVM / Decision Tree below

```python
##LR
lr = LogisticRegression(penalty='l1',solver='liblinear',C=0.01,random_state=1)
lr.fit(X_train_std, y_train)
lr2 = LogisticRegression(penalty='l1',solver='liblinear',C=10,random_state=1)
lr2.fit(X_train_std, y_train)###try several C, and finally set it as 0.01

print('C=0.01,Training accuracy:', lr.score(X_train_std, y_train))
print('C=0.01,Test accuracy:', lr.score(X_test_std, y_test))
print('C=10,Training accuracy:', lr2.score(X_train_std, y_train))
print('C=10,Test accuracy:', lr2.score(X_test_std, y_test))
```

```python
## SVM
svm = SVC(kernel='rbf', random_state=1, gamma=0.1, C=10.0)
svm.fit(X_train_std, y_train)
svm2 = SVC(kernel='rbf', C=1.0, gamma=0.01,random_state=1)
svm2.fit(X_train_std, y_train)###try several C and gamma, and finally set C=10, gamma=0.1

print('C=10 gamma=0.1 Training accuracy:', svm.score(X_train_std, y_train))
print('C=10 gamma=0.1 Test accuracy:', svm.score(X_test_std, y_test))
print('C=1,gmma=0.01 Training accuracy:', svm2.score(X_train_std, y_train))
print('C=1,gmma=0.01 Test accuracy:', svm2.score(X_test_std, y_test))
```

```python
## Decision Tree
tree = DecisionTreeClassifier(criterion='gini', 
                              max_depth=4, 
                              random_state=1)
tree.fit(X_train_std, y_train)

tree2 = DecisionTreeClassifier(criterion='gini', 
                              max_depth=5, 
                              random_state=1)
tree2.fit(X_train_std, y_train)###try several max_depth, and finally set it as 4

print('max_depth=4 Training accuracy:', tree.score(X_train_std, y_train))
print('max_depth=4 Test accuracy:', tree.score(X_test_std, y_test))
print('max_depth=5 Training accuracy:', tree2.score(X_train_std, y_train))
print('max_depth=5 Test accuracy:', tree2.score(X_test_std, y_test))
```

## Visualize the classification

```python
from matplotlib.colors import ListedColormap
import matplotlib.pyplot as plt

def plot_decision_regions(X, y, classifier, test_idx=None, resolution=0.02):

    # setup marker generator and color map
    markers = ('s', 'x', 'o', '^', 'v')
    colors = ('red', 'blue', 'lightgreen', 'gray', 'cyan')
    cmap = ListedColormap(colors[:len(np.unique(y))])

    # plot the decision surface
    x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution),
                           np.arange(x2_min, x2_max, resolution))
    Z = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
    Z = Z.reshape(xx1.shape)
    plt.contourf(xx1, xx2, Z, alpha=0.3, cmap=cmap)
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())

    for idx, cl in enumerate(np.unique(y)):
        plt.scatter(x=X[y == cl, 0], 
                    y=X[y == cl, 1],
                    alpha=0.8, 
                    c=colors[idx],
                    marker=markers[idx], 
                    label=cl, 
                    edgecolor='black')

    # highlight test samples
    if test_idx:
        # plot all samples
        X_test, y_test = X[test_idx, :], y[test_idx]

        plt.scatter(X_test[:, 0],
                    X_test[:, 1],
                    c='',
                    edgecolor='black',
                    alpha=1.0,
                    linewidth=1,
                    marker='o',
                    s=100, 
                    label='test set')
```

```python
##LR
X_combined_std = np.vstack((X_train_std, X_test_std))
y_combined = np.hstack((y_train, y_test))

plot_decision_regions(X=X_combined_std, y=y_combined,
                      classifier=lr, test_idx=range(6854,9792))
plt.xlabel('X1')
plt.ylabel('X38')
plt.legend(loc='upper left')

plt.tight_layout()
#plt.savefig('images/03_01.png', dpi=300)
plt.show()
```

```python
##SVM it will take a long time to show this picture
plot_decision_regions(X_combined_std, 
                      y_combined,
                      classifier=svm, 
                      test_idx=range(6854,9792))
plt.xlabel('X1')
plt.ylabel('X38')
plt.legend(loc='upper left')
plt.tight_layout()
#plt.savefig('images/03_11.png', dpi=300)
plt.show()
```

```python
## Decition Tree
plot_decision_regions(X_combined_std, y_combined, 
                      classifier=tree, test_idx=range(6854,9792))

plt.xlabel('X1')
plt.ylabel('X38')
plt.legend(loc='upper left')
plt.tight_layout()
#plt.savefig('images/03_20.png', dpi=300)
plt.show()
```
